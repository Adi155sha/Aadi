```python
results = []
for i in df['quality']:
    if i >= 1 and i <= 3:
        results.append('1')
    elif i >= 4 and i <= 7:
        results.append('2')
    elif i >= 8 and i <= 10:
        results.append('3')
df['Result'] = results
```

# Modelling

```python
features = ['citric acid','alcohol','residual sugar','pH','free sulfur dioxide','chlorides']  #input values

x = df[features]
y = df['quality']            
```

using train_test_split function to split the dataset.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.10,shuffle=True ,random_state=3)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

Applying KNN classification on "quality" column of dataset.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7,p=2)
knn.fit(x_train,y_train)

#predicting on the test set
predict_knn=knn.predict(x_test)
print(knn.score(x_test,y_test))
```

Finding the Accuracy

```python
print("fraction of correct predictions is =",(y_test==predict_knn).sum(),"/",(y_test!=predict_knn).sum()+(y_test==predict_knn).sum())
```

```python
print("the accuracy is",100*knn.score(x_test,y_test),"%")
```
